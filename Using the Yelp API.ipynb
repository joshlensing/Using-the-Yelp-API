## Imports

```python
import pandas as pd
```

```python
# use keys to login to yelpapi
import json
with open('/Users/joshl/.secret/yelp_api.json') as f:
    login = json.load(f)
login.keys()
```

```python
# import the YelpAPI Class
from yelpapi import YelpAPI
# Create an instance with your key
yelp_api = YelpAPI(login['api-key'], timeout_s=5.0)
yelp_api
```

I want to search for restaurants that serve clam chowder in Boston, since it is a New England favorite.

```python
# use our yelp_api variable's search_query method to perform our API call
search_results = yelp_api.search_query(location='Boston, MA',
                                       term='Clam Chowder')
print(type(search_results))
search_results.keys()
```

```python
# I want to see how many search results are returned for the location and food
search_results['total']
```

There are 1400 business results for "clam chowder." Yelp will only return 20 when using the search results for businesses since it can only display 20 per page. I want to be able to display more than that, so I will use offset to display more.

```python
bus = pd.DataFrame(search_results['businesses'])
bus.head() 
```

```python
# add offset to our original api call
search_results_1 = yelp_api.search_query(location='Boston, MA',
                                       term='Clam Chowder',
                                       offset = 20)
```

Now I will create a Pandas DataFrame for the offset search results.

```python
bus20 = pd.DataFrame(search_results_1['businesses'])
bus20.head()
```

I can now combine both dataframes to return a dataframe with a higher number of results. This will be done by concatenating the dataframes.

```python
## concatenate the previous results and new results. 
businesses = pd.concat([bus, bus20],
                      ignore_index=True)
display(businesses.head(3), businesses.tail(3))
```

Now I want to save the businesses dataframe to a records-oriented JSON file.

```python
businesses.to_json(orient='records')
```
